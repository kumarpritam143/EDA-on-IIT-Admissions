```python
# Group data by Field of Study and calculate average fees and discounts
grouped_data = p.groupby('Field of Study').agg({'Fees': 'mean', 'Discount on Fees': 'mean'})

# Plotting bar chart for average fees and discounts
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', ax=plt.gca())
plt.title('Average Fees and Discounts by Field of Study')
plt.xlabel('Field of Study')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
graduation_counts = p.groupby('Expected Year of Graduation').size()

# Create a bar plot for the counts
plt.figure(figsize=(10, 6))
graduation_counts.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Number of Students Graduating Each Year')
plt.xlabel('Year of Graduation')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Count the number of students for each specialization
specialization_counts = p['Specialization'].value_counts()

# Create a bar plot for the counts
plt.figure(figsize=(10, 6))
specialization_counts.plot(kind='bar', color='orange', alpha=0.7)
plt.title('Number of Students for Each Specialization')
plt.xlabel('Specialization')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter students studying Computer Science
cs_students = p[p['Field of Study'] == 'Computer Science']

# Count the number of students for each specialization
specialization_counts = cs_students['Specialization'].value_counts()

# Create a bar plot for the counts
plt.figure(figsize=(10, 6))
specialization_counts.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Specializations of Computer Science Students')
plt.xlabel('Specialization')
plt.ylabel('Number of Students')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter students expected to graduate in 2020
graduates_2020 = p[p['Expected Year of Graduation'] == 2020]

# Count the number of students for each semester
semester_counts = graduates_2020['Current Semester'].value_counts()

# Create a bar plot for the counts
plt.figure(figsize=(10, 6))
semester_counts.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Semester Distribution of Students Graduating in 2020')
plt.xlabel('Semester')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Filter students admitted in 2020
admission_2020 = p[p['Year of Admission'] == 2020]

# Count the number of students for each field of study
field_counts = admission_2020['Field of Study'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(field_counts, labels=field_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Students Admitted in 2020 by Field of Study')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
# Filter students with date of birth in 2000
dob_2000 = p[p['Date of Birth'].str.startswith('2000')]

# Count the number of students for each field of study
field_counts = dob_2000['Field of Study'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
field_counts.plot(kind='bar', color='purple', alpha=0.7)
plt.title('Field of Study of Students with Date of Birth in 2000')
plt.xlabel('Field of Study')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
